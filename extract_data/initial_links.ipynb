```python
names = [hack["name"] for year in hackathons.values() for hack in year]
len(names)
```

```python
# Create a list to store all the base URLs for the project galleries
base_urls = []

# Iterate through each year's list of hackathons
for year_events in hackathons.values():
    # Iterate through each event in the list
    for event in year_events:
        # Get the base link and remove any trailing slash
        base_link = event['link'].rstrip('/')
        # Create the project gallery URL and append it to our list
        project_gallery_url = f"{base_link}/project-gallery?"
        base_urls.append(project_gallery_url)

# Print the final list of base URLs
print(len(base_urls))
```

```python
print(names[42],base_urls[42])
print(names[12],base_urls[12])
```

```python
submissions = []
count = 1
for url in base_urls:
    response = req.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    pagination_div = soup.find("div", class_="pagination-info")
    numbers = pagination_div.find("span", class_="items_info").find_all("b")
    total = numbers[1].text.strip()
    submissions.append(total)
    print("Percentage",((count/50)))
    count += 1
```

```python
submissions = list(map(int, submissions))
submissions
```

```python
all_project_links = {}
j = 0
for base_url, sub_count in zip(base_urls, submissions):
    pages = math.ceil(sub_count / 24)  # round up to get number of pages
    project_links = [f"{base_url}page={i}" for i in range(1, pages + 1)]
    all_project_links[names[j]] = project_links
    j += 1
# Example: print links for HackHarvard
all_project_links
```

```python
def scrape_all_hackathons(hackathon_dict, output_file="all_hackathons.csv"):
    all_project_data = []  # collect data from all hackathons

    for hackathon_name, links in hackathon_dict.items():
        print(f"🔎 Scraping {hackathon_name}...")
        for url in links:
            response = req.get(url)
            soup = BeautifulSoup(response.text, "html.parser")
            links_soup = soup.find_all("a", class_="block-wrapper-link fade link-to-software")

            for a in links_soup:
                href = a["href"]
                title_tag = a.find("h5")
                title = title_tag.get_text(strip=True) if title_tag else ""
                winner_tag = a.find("img", class_="winner")
                is_winner = True if winner_tag else False

                all_project_data.append({
                    "hackathon_name": hackathon_name,
                    "project_name": title,
                    "project_link": href,
                    "winner": is_winner
                })

        print(f"✅ Finished {hackathon_name}, total collected so far: {len(all_project_data)}")

    # save once at the end
    df = pd.DataFrame(all_project_data)
    df.to_csv(output_file, index=False)
    print(f"🎉 Saved {len(all_project_data)} projects from {len(hackathon_dict)} hackathons to {output_file}")
```
